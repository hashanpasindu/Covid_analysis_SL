```python
from statsmodels.tsa.stattools import adfuller
```

```python

adfuller(result2['new_cases'], autolag='AIC')
```

```python
import numpy as np
```

```python
result2['new_cases'] = result2['new_cases'].apply(lambda x: np.log(x))
```

```python

from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition = seasonal_decompose(result2['new_cases'], freq = 180)
```

```python

result2.loc[:,'trend'] = decomposition.trend
result2.loc[:, 'seasonal'] = decomposition.seasonal
result2.loc[:, 'residual'] = decomposition.resid
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (15, 5), sharex = True)
ax1.plot(result2['trend'], label='Original')
ax1.legend(loc='best')
ax1.tick_params(axis = 'x', rotation = 45)

ax2.plot(result2['seasonal'], label='Seasonal')
ax2.legend(loc='best')
ax2.tick_params(axis = 'x', rotation = 45)

ax3.plot(result2['residual'], label='Residual')
ax3.legend(loc='best')
ax3.tick_params(axis = 'x', rotation = 45)
```
